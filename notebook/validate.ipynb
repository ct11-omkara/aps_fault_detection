```python
#The null hypothesis       : the two distributions are identical, F(x) = G(x) for all x
#The alternative hypothesis: the two distributions are not identical, , F(x) != G(x) for all x

# If my pvalue is more than 0.05, then we will accept null hypothesis...
# If my pvalue is less than 0.05, then we will reject null hypothesis...
```

```python
import numpy as np
from scipy import stats
rng = np.random.default_rng()
sample1 = stats.uniform.rvs(size=100, random_state=rng)
sample2 = stats.norm.rvs(size=110, random_state=rng)
response = stats.ks_2samp(sample1, sample2)
```

```python
sample2
```

```python
# If my pvalue is less than 0.05, then we will reject null hypothesis, it means trained data and new incomming training data both have not same data distribution....
# So, we have to retrain our model with new incomming train data....
response.pvalue, round(response.pvalue, 3)
```
